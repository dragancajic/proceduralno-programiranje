```python
# Enumerativni algoritmi (Sudoku)
# Zadatak 
# Problem trgovačkog putnika / slika u prilogu
# TSP ruta je 1 -> 2 -> 4 -> 3 -> 1.  10 + 25 + 30 + 15 = 80.

V = 4
answer = []
 
# Funkcija koja pronalazi minimalnu putanju
def tsp(graph, v, currPos, n, count, cost):
 
    # Ako smo došli do posljednjeg čvora 
    # i ako je on povezan sa početnim čvorom
    if (count == n and graph[currPos][0]):
        answer.append(cost + graph[currPos][0])
        return
 
    # BACKTRACKING STEP
    
    for i in range(n):
        if (v[i] == False and graph[currPos][i]):
             
            # Označi kao posjećen
            v[i] = True
            tsp(graph, v, i, n, count + 1,
                cost + graph[currPos][i])
            
             
            # Označi kao neposjećen
            v[i] = False
 

 
# n je broj čvorova
if __name__ == '__main__':
    n = 4
    graph= [[ 0, 10, 15, 20 ],
            [ 10, 0, 35, 25 ],
            [ 15, 35, 0, 30 ],
            [ 20, 25, 30, 0 ]]
 
    # Boolean lista koja provjerava da li je čvor
    # posjećen ili ne
    v = [False for i in range(n)]
     
    # Označimo da je prvi (nulti) čvor posjećen
    v[0] = True
 

    tsp(graph, v, 0, n, 1, 0)
 
    
    print(answer)
    print(min(answer))
# 1 -> 2 -> 3 -> 4 - 95 
# 1 -> 2 -> 4 -> 3 - 80
```

```python
# Zadatak 
# Napisati program koji ispisuje sve permutacije jednog stringa

"""
Npr. za string ABC permutacije su: 
ABC ACB BAC BCA CBA CAB
"""

 
def toString(List):
    return ''.join(List)
 
# Funkcija za printanje permutacija stringa
# Funkcija uzima tri parametra
# 1. String
# 2. Početni indeks stringa
# 3. Krajnji indeks stringa.
def permute(a, l, r):
    if l==r:
        print (toString(a))
    else:
        for i in range(l,r):
            a[l], a[i] = a[i], a[l]
           
            permute(a, l+1, r) # rekurzija
            
            a[l], a[i] = a[i], a[l] # backtrack

string = "ABC"
n = len(string)
a = list(string)
permute(a, 0, n)
```

```python
# Zadatak 
# Napisati program koji ispisuje sve podskupove datog skupa

"""
Npr. za skup a = {1, 2, 3}
Izlaz je: // ovako oznacimo prazan skup 
         1
         1 2
         1 2 3
         1 3
         2
         2 3
         3
"""
# U listi u svakom koraku imamo dva
# izbora za svaki element - ili cemo ga ignorisati 
# ili ćemo ga uključiti u podskup
def subsetsUtil(A, subset, index):
    print(*subset)
    for i in range(index, len(A)):
         
        # uključuje A[i] u podskup
        subset.append(A[i])
         
        # prelazak na sljedeći element
        subsetsUtil(A, subset, i + 1)
         
        # isključivanje A[i] iz podskupa
        # triggers backtracking
        subset.pop()
    return
 
# vraća podskup od A.
def subsets(A):
    #global res
    subset = []
    index = 0
    subsetsUtil(A, subset, index)
     

array = [1, 2, 3]
 

# O(2 ^ (n))
# zato što u svakom koraku imamo dva izbora

subsets(array)
 
```

```python
# Zadatak 
# Lavirint - slika u prilogu kao ilustracija

""" Data je binarna matrica dimenzija nxn. Matrica [0][0] je početna pozicija a cilj je na poziciji
matrica [n-1][n-1].
Potrebno je od startne pozicije doći do cilja ali da se pritom mogu koristiti samo koraci 
naprijed (desno) i dolje.
Takođe, matrica je popunjena sa nulama i jedinicama gdje nule znače zabranjenu poziciju, tj. 
da taj kvadrat ne može biti sastavni dio putanje dok jedinica označava kvadrat koji može biti dio
putanje od starta do cilja.

Npr.
{1, 0, 0, 0}
{1, 1, 0, 1}
{0, 1, 0, 0}
{1, 1, 1, 1}

"""
# Rat in a Maze

 
# Veličina lavirinta
N = 4
 
# Pomoćna funkcija za ispis rješenja matrice
def printSolution( sol ):
     
    for i in sol:
        for j in i:
            print(str(j) + " ", end ="")
        print("")

# Pomoćna funkcija za provjeru da li su x, y odgovarajući indeksi
# za NxN lavirint
def isSafe( maze, x, y ):
     
    if x >= 0 and x < N and y >= 0 and y < N and maze[x][y] == 1:
        return True
     
    return False
 
""" Ova funkcija rješava problem koristeći Backtracking.
    Funkcija vraća False ako nije moguće naći odgovarajući put
    inače, vraća True i ispisuje put od starta do cilja. 
    Moguće je da  postoji više rješenja, međutim funkcija vraća samo
    jedno od njih."""
def solveMaze( maze ):
     
    # Kreiranje 4 * 4 matrice - popunimo je nulama
    sol = [ [ 0 for j in range(4) ] for i in range(4) ]
     
    if solveMazeUtil(maze, 0, 0, sol) == False:
        print("Rješenje ne postoji!");
        return False
     
    printSolution(sol)
    return True

# Rekurzivna pomoćna funkcija za rješavanje problema
def solveMazeUtil(maze, x, y, sol):
     
    # ako za cilj odgovaraju x, y  vrati True
    if x == N - 1 and y == N - 1 and maze[x][y]== 1:
        sol[x][y] = 1
        return True
         
    # Provjera za [x][y] da li se mogu uključiti u putanju 
    if isSafe(maze, x, y) == True:
        # Provjera da li je ta ćelija već uključena u put   
        if sol[x][y] == 1:
            return False
           
        # označi x i y kao dio puta
        sol[x][y] = 1
         
        # Pomjeranje naprijed po x osi - desno
        if solveMazeUtil(maze, x + 1, y, sol) == True:
            return True
             
        # Ako pomjeranje desno nije dalo rezultat
        # onda se pomjeriti prema dolje, tj. pomjera se po y osi
        if solveMazeUtil(maze, x, y + 1, sol) == True:
            return True
           
        # Ako pomjeranje po y osi nije dalo rješenje
        # onda se vraćamo nazad po x pravcu
        if solveMazeUtil(maze, x - 1, y, sol) == True:
            return True
             
        # Ako vraćanjem nazad ne nađemo rješenje 
        # potrebno je vratiti se nazad i po y osi
        if solveMazeUtil(maze, x, y - 1, sol) == True:
            return True
         
        # Ako se ne pronalazi nijedno rješenje
        # BACKTRACK: x, y nisu dio rješenja odnosno putanje
        sol[x][y] = 0
        return False


if __name__ == "__main__":
   
    maze = [ [1, 0, 0, 0],
             [1, 1, 0, 1],
             [0, 1, 0, 0],
             [1, 1, 1, 1] ]
              
    solveMaze(maze)
 
```

```python
# Pomoćni zadatak
# Napisati sve moguće putanje u matrici formata nxm (početak i cilj kao i način kretanja
# isti kao u primjeru iznad)

"""
Primjer  

Ulaz  :  1 2 3
         4 5 6
        
Izlaz :  1 4 5 6
         1 2 5 6
         1 2 3 6

Ulaz:    1 2 
         3 4
         
Izlaz:   1 2 4
         1 3 4
"""

# Program za ispis svih mogućih putanja od gornjeg lijevog do donjeg desnog elementa u mxn matrici

 

def printAllPathsUtil(mat, i, j, m, n, path, pi):
 
    # Stići do cilja jedino uz opciju kretanja udesno
    
    if (i == m - 1):
        for k in range(j, n):
            path[pi + k - j] = mat[i][k]
 
        for l in range(pi + n - j):
            print(path[l], end = " ")
        print()
        return
 
    # # Stići do cilja jedino uz opciju kretanja nadole
    if (j == n - 1):
 
        for k in range(i, m):
            path[pi + k - i] = mat[k][j]
 
        for l in range(pi + m - i):
            print(path[l], end = " ")
        print()
        return
 
    # Dodati trenutnu poziciju
    path[pi] = mat[i][j]
 
    # Ispis svih putanja kretanjem nadole
    printAllPathsUtil(mat, i + 1, j, m, n, path, pi + 1)
 
    # Ispis svih putanja kretanjem udesno
    printAllPathsUtil(mat, i, j + 1, m, n, path, pi + 1)
 
    # Printanje svih puteva kretanjem dijagonalno - za kvadratne matrice
    # printAllPathsUtil(mat, i+1, j+1, m, n, path, pi + 1);

# Glavna funkcija koja ispisuje sve putanje 

def printAllPaths(mat, m, n):
 
    path = [0 for i in range(m + n)]
    printAllPathsUtil(mat, 0, 0, m, n, path, 0)


mat = [[1, 2, 3],
       [4, 5, 6]]
 
printAllPaths(mat, 2, 3)
```

```python
# Zadatak 

# Napisati program koji pronalazi sve podskupove datog stringa ali tako da budu
# leksikografski poredani

"""
Npr.
Ulaz: cab
Izlaz: a ab abc ac b bc c
"""

 

def permuteRec(string, n, index = -1, curr = ""):
 
    # bazni slučaj
    if index == n:
        return
 
    if len(curr) > 0:
        print(curr)
 
    for i in range(index + 1, n):
        curr += string[i]
        permuteRec(string, n, i, curr) #rekurzija
 
        # backtracking
        curr = curr[:len(curr) - 1]

def powerSet(string):
    string = ''.join(sorted(string))
    permuteRec(string, len(string))

if __name__ == "__main__":
    string = "cab"
    powerSet(string)
 
```

```python
# Zadatak  - dp preći kasnije ovaj zadatak
#Subset Sum Problem - detaljnije na linku: https://favtutor.com/blogs/dynamic-programming
"""
# Objašnjenja - https://www.geeksforgeeks.org/subset-sum-problem-dp-25/
https://towardsdatascience.com/how-to-find-all-solutions-to-the-subset-sum-problem-597f77677e45
Pitanje u vezi sa ovim problemom se često pojavljuje na inetrvjuima za posao u Microsoft-u.

Subset sum problem je problem pronalaženja podskupa datog skupa brojeva gdje je suma elemenata 
poskupa jednaka određenom target broju.
"""
def subSum(arr,target):
    n = len(arr)
    dp = [[None]*(target+1) for i in range(n+1)]
    # print(dp)
    # Inicijalizacija liste
    # Ako je broj elemenata liste 0 tada je nije moguće odrediti Target sum
    
    for i in range(target+1):
        dp[0][i] = False
    for j in range(n+1):
        dp[j][0] = True
    #print(dp)
    # Element može biti izabran jedino ako je suma manja od arr[i] inače ga ne uključujemo
    for i in range(1,n+1):
        for j in range(target+1):
            if arr[i-1] <= j:
                dp[i][j] = dp[i-1][j-arr[i-1]] or dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j]

    return dp[n][target]

if __name__ == '__main__':
    arr = [0, 3, 2, 7, 1]
    target = 6
    exists =  subSum(arr,target)
    print('Podskup sa datom target sumom postoji : ', exists)
```

```python
# Rekurzivno - za domaći
  

def isSubsetSum(set, n, sum) :
    
    # Bazni slučaj
    if (sum == 0) :
        return True
    if (n == 0 and sum != 0) :
        return False
   
    # Ako je element veći od sume ignoriši ga
    if (set[n - 1] > sum) :
        return isSubsetSum(set, n - 1, sum);
   

    
    # (a) uključivanje posljednjeg elementa
    # (b) isključivanje posljednjeg elementa
    return isSubsetSum(set, n-1, sum) or isSubsetSum(set, n-1, sum-set[n-1])
      
      

set = [3, 34, 4, 12, 5, 2]
sum = 9
n = len(set)
if (isSubsetSum(set, n, sum) == True) :
    print("Pronađen je podskup sa datom sumom")
else :
    print("Nije pronađen podskup sa datom sumom")
      
```

```python
# Zadatak 

"""
Neka je data lista pozitivnih cijelih brojeva i suma x.
Potrebno je napisati program koji pronalazi sve jedinstvene kombinacije brojeva
iz liste kojima je suma jednaka X. Brojevi mogu biti izvučeni iz liste neograničen
broj puta. Kombinacije elemenata potrebno je ispisati u rastućem redoslijedu.
Kombinacije moraju biti sortirane u opadajućem redoslijedu tako da kombinacija
sa najmanjim prvim elementom mora da bude isprintana prva. Ukoliko nema kombinacija potrebno 
je ispisati "Prazan skup".

Primjer: 

Ulaz: lista = [2, 4, 6, 8] 
            x = 8
Izlaz:   [2, 2, 2, 2]
         [2, 2, 4]
         [2, 6]
         [4, 4]
         [8]

"""
def combinationSum(arr, sum):
    ans = []
    temp = []
 
    
    arr = sorted(list(set(arr)))
    findNumbers(ans, arr, temp, sum, 0)
    return ans
 
def findNumbers(ans, arr, temp, sum, index):
     
    if(sum == 0):
         
        
        ans.append(list(temp))
        return
       
    
    for i in range(index, len(arr)):
 
    
        if(sum - arr[i]) >= 0:
 
            
            temp.append(arr[i])
            findNumbers(ans, arr, temp, sum-arr[i], i)
 
            # brisanje elementa iz liste (backtracking)
            temp.remove(arr[i])
 
 

arr = [2, 4, 6, 8]
sum = 8
ans = combinationSum(arr, sum)
 
    
if len(ans) <= 0:
    print("Prazan skup")
     

for i in range(len(ans)):
 
    print("(", end=' ')
    for j in range(len(ans[i])):
        print(str(ans[i][j])+" ", end=' ')
    print(")", end=' ')
 
```
