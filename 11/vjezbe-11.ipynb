```python
# Pohlepni algoritmi

# Zadatak 1 

"""
Napisati program koji pronalazi najmanji broj čija je suma cifara
jednaka N i koji je
djeljiv sa 10**N.

Input: N = 5
Output: 500000

Input: N = 20
Output: 29900000000000000000000

"""
 
import math
def digitsNum(N):
 
    # Ako je N=0 vratiće se string 0
    if (N == 0) :
        print("0", end = "")
     
    # Ako N daje ostatak pri dijeljenju sa 9
    # isprintati ostatak
    if (N % 9 != 0):
        print (N % 9, end ="")
     
    # Ispisati 9 N/9 puta
    for i in range( 1, int(N / 9) + 1) :
        print("9", end = "")
     
    # Pridružiti N nula broju 
    # kako bi bio djeljiv sa 10**N
    for i in range(1, N + 1) :
        print("0", end = "")
     
    print()
 
 

N = 20
print("Broj je: ",end="")
digitsNum(N)
 
```

```python
# Zadatak 2

"""
Bin Packing Problem (Minimizovati broj korištenih korpi)
Dato je n stavki različite težine i korpe svaka kapaciteta c.
Potrebno je napisati program koji pronalazi najmanji broj korpi tako da
sve stavke budu pokupljene.


Input:  težine = [4, 8, 1, 4, 2, 1]
        Kapacitet korpe c = 10
Output: 2


Potrebne su minimum dvije korpe da se smjeste sve stvari 
Prva korpa sadrži {4, 4, 2}, druga korpa sadrži {8, 1, 1}

Još par primjera

Input:  weight[] = {9, 8, 2, 2, 5, 4}
        Bin Capacity c = 10
Output: 4

Input:  weight[] = {2, 5, 4, 7, 1, 3, 8}; 
        Bin Capacity c = 10
Output: 3
"""

def nextfit(weight, c):
    res = 0
    rem = c
    for i in range(len(weight)):
        if rem >= weight[i]:
            rem = rem - weight[i]
        else:
            res += 1
            rem = c - weight[i]
    return res
 

weight = [2, 5, 4, 7, 1, 3, 8]
c = 10
 
print("Broj potrebnih korpi - Next Fit :",
                           nextfit(weight, c))
 
    
    

# First Fit algorithm - za pronalazak broja korpi
 

def firstFit(weight, n, c):
     
     
    # Broj korpi na početku postavljamo na 0
    res = 0
     
    # Kreiranje liste za skladištenje preostalog prostora u korpi
    # najviše može biti n korpi
    bin_rem = [0]*n
    #print(bin_rem)
    # Stavljaj stvari jednu po jednu
    for i in range(n):
       
        # Pronađi prvu korpu koja je prikladna za
        # weight[i]
        j = 0
        while( j < res):
            if (bin_rem[j] >= weight[i]):
                bin_rem[j] = bin_rem[j] - weight[i]
                break
            j+=1
             
        # Ako ne postoji korpa koja može da smjesti weight[i]
        if (j == res):
            bin_rem[res] = c - weight[i]
            res= res+1
    return  res, bin_rem    

weight = [2, 5, 4, 7, 1, 3, 8]
c = 10
n = len(weight)
print("Broj potrebnih korpi - First Fit : ",firstFit(weight, n, c))
 
```

```python
# Zadatak 3
# Pronaći najmanji broj ako je dat broj
# njegovih cifara i suma cifara

"""
Primjer:

Input  : s = 9, m = 2
Output : 18

Postoji još mogućnosti kao što su 45, 54, 90, ali najmanji je 18.

Input  : s = 20, m = 3
Output : 299
"""

 
def findSmallest(m,s):
 
    # Ako je suma cifara 0,
    # Moguće je jedino da je broj cifara 1
    if (s == 0):         
        if(m == 1) :
              print("Najmanji broj je 0")
        else :
              print("Nemoguće je pronaći traženi broj")
        return
  
    # Suma veća od maksimalne moguće sume    
    if (s > 9*m):
        print("Nemoguće je pronaći traženi broj")
        return
  
    # Kreiramo listu za skladištenje cifara
    
    res=[0 for i in range(m+1)]
    s -= 1  
    # Popuniti preostalih m-1 cifara
    # sa desne na lijevu stranu
    for i in range(m-1,0,-1):     
        # Ako je suma veća od 9,
        # cifra mora biti 9.
        if (s > 9):
         
            res[i] = 9
            s -= 9     
        else:
            res[i] = s
            s=0
  
    res[0] += s + 1                    
    print("Najmanji traženi broj je: ",end="")
    for i in range(m):
        print(res[i],end="")

findSmallest(3, 20)
 
```

```python
# Zadatak 4

"""
Rasporediti miševe u rupe ako postoji N miševa i N rupa koje su smještene na
pravoj liniji.
U svaku rupu može se smjestiti samo jedan miš. 
Miš može stajati na svojoj poziciji,
može se pomaći jedan korak desno sa x na x+1 ili se pomaći jedan korak
lijevo sa x na x-1.
Svako pomjeranje traje jedan minut. 
Rasporediti miševe u rupe tako da vrijeme raspoređivanja bude minimalno. 

Primjeri: 

Input : Pozicije miševa su:
          4 -4 2
        Pozicije rupa su:
          4 0 5
Output :  4

Objašnjenje
Miš na poziciji x = 4 smješta se u rupu na poziciji x = 4
Vrijeme koje je potrebno za to je 0 minuta 

Miš na poziciji x = -4 smješta se u rupu na poziciji x = 0
Vrijeme koje je potrebno za to je 4 minuta 

Miš na poziciji x = 2 smješta se u rupu na poziciji x = 5
Vrijeme koje je potrebno za to je 3 minuta 

Miševi se mogu kretati nezavisno jedan od drugog pa je rješenje zadatka:
za 4 minute svi miševi će biti raspoređeni u rupe.

Još jedan primjer:

Input :  Pozicije miševa su:
        -10, -79, -79, 67, 93, -85, -28, -94 
         Pozicije rupa su:
         -2, 9, 69, 25, -31, 23, 50, 78 
Output : 102

"""

def assignHole(mices, holes, n, m):
     
    # Bazni slučaj
    # Broj miševa i rupa mora biti isti
    if (n != m):
        return -1
     
    # Sortirati liste
    mices.sort()
    holes.sort()
     
    # Pronaći najveću razliku između i-tog miša i i-te rupe 
    Max = 0
     
    for i in range(n):
        if (Max < abs(mices[i] - holes[i])):
            Max = abs(mices[i] - holes[i])
     
    return Max
     
  
 
# Pozicije miševa
mices = [ 4, -4, 2 ]
 
# Pozicije rupa
holes = [ 4, 0, 5 ]
n = len(mices)
m = len(holes)

minTime = assignHole(mices, holes, n, m)
 
print("Posljednji miš će ući u rupu za:", minTime, "minut(a/e)")
```

```python
# Zadatak 5 

"""
Fractional Knapsack Problem - Problem ranca

Neka su date težine i vrijednosti za n stvari.
Treba da smjestimo stvari u ranac kapaciteta W ali tako da imamo maksimalnu vrijednost u rancu.

Parovi (value, weight)  
arr[] = {{60, 10}, {100, 20}, {120, 30}} 
Kapacitet ruksaka, W = 50; 

Output: 
Maksimalna moguća vrijednost = 240 
ako se pakuje stvar težine 10 i 20 kg i 2/3 od 30 kg. 
Slijedi da je ukupna cijena 60+100+(2/3)(120) = 240

"""
"""
weights = [6, 7, 3]
values = [110, 120, 2]
capacity = 10
"""
def solve(weights, values, capacity):
    res = 0
    for pair in sorted(zip(weights, values), key=lambda x: - x[1]/x[0]):
        print(pair)
        if not bool(capacity):
            break
        if pair[0] > capacity:
            res += int(pair[1] / (pair[0] / capacity))
            capacity = 0
        elif pair[0] <= capacity:
            res += pair[1]
            capacity -= pair[0]
    return int(res)


weights = [10, 40, 20, 30]
values = [60, 40, 100, 120]
capacity = 50
print(solve(weights, values, capacity))

```

```python
# Zadatak 6

"""
Minimalan broj zamjena uglastih zagrada
Neka je dati string od 2N karaktera koji ima N ‘[‘ zagrada i N ‘]’ zagrada.
String se smatra balansiranim ako se može predstaviti na sljedeći način:
S2[S1] gdje su S1 i S2 balansirani stringovi
Potrebno je izračunati minimalni broj zamjena koje je neophodno
uraditi da bi string bio balansiran.

Input  : []][][
Output : 2
Prva zamjena: Pozicije 3 i 4
[][]][
Druga zamjena: Pozicije 5 i 6
[][][]

Input  : [[][]]
Output : 0
String je već balansiran.
"""

def swapCount(s):
 
    # Čuvati indekse pojavljivanja '['
    pos = []
 
    for i in range(len(s)):
        if(s[i] == '['):
            pos.append(i)
 
    # broj zagrada '['        
    count = 0
     
    # Pozicija sljedeće zagrade '[' u pos
    p = 0   
    sum = 0       
    s = list(s)
     
    for i in range(len(s)):
        if(s[i] == '['):
            count += 1
            p += 1
        elif(s[i] == ']'):
            count -= 1
 
       
        if(count < 0):
            sum += pos[p] - i
            s[i], s[pos[p]] = (s[pos[p]],
                               s[i])
            p += 1
            count = 1
    return sum
 

s = "[]][]["
print(swapCount(s))
 
#s = "[[]]][][]["
#print(swapCount(s))
 
    

```

```python
# Drugi način

def swapCount(s):   
    chars = s  
    # Čuvamo ukupan broj oba tipa zagrada
    countLeft = 0
    countRight = 0
    swap = 0
    imbalance = 0
     
    for i in range(len(chars)):
        if chars[i] == '[':
            countLeft += 1
            if imbalance > 0:
                swap += imbalance
                imbalance -= 1
                 
        elif chars[i] == ']':
            countRight += 1
            imbalance = (countRight - countLeft)
 
    return swap
 

s = "[]][][";
print(swapCount(s))
 
s = "[[][]]";
print(swapCount(s))
 
```

```python
# Zadatak 8
# Problem razmjenjivanja novca. Imamo X KM-ova i na
# rapolaganju apoene od x1, x2, . . . , xk KM-ova
# negoraničenog broja.
# Kako razmijeniti X KM-ova tako da dobijemo što manji broj
# novčića.
# Objašnjenje dato na: https://www.geeksforgeeks.org/greedy-algorithm-to-find-minimum-number-of-coins/  

def findMin(V):
      
    # Svi apoeni
    deno = [1, 2, 5, 10, 20, 50, 
            100, 500, 1000]
    n = len(deno)
    ans = []
    i = n - 1
    while(i >= 0):
          
        while (V >= deno[i]):
            V -= deno[i]
            ans.append(deno[i])
  
        i -= 1  
    for i in range(len(ans)):
        print(ans[i], end = " ")
if __name__ == '__main__':
    n = 93
    print("Rješenje - ",
          "razmjena za ", n, "KM: ", end = "")
    findMin(n)
      
```
