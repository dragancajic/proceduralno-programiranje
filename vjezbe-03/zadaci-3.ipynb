```python
# Zadatak 1

"""Sa tastature se unosi broj koji ima 4 ili više cifara.
Provjeriti da li je proizvod prostih faktora unesenog broja veći od proizvoda
prostih faktora broja koji se dobije tako što se unesenom broju obrnu cifre na parnim pozicijama.
Npr. uneseni broj je 15879249, a broj dobijen na opisani način je 19829745.
"""

def cifre_broja(n):
    lista=[]
    while n > 0:
        c = n % 10
        n //= 10
        lista.insert(0, c)
    return lista

def novi_broj(n):
    '''
    računa broj koji se dobije tako što mu se obrnu cifre na parnim pozicijama,
    npr. 15879249 -> 19829745.
    '''
    cifre_n = cifre_broja(n)
    cifre_neparne_pozicije = cifre_n[::2]
    cifre_parne_pozicije = cifre_n[1::2]  # √
    # obrnuta lista cifara na parnim pozicijama, možemo koristiti cifre_parne_pozicije.reverse()
    cifre_parne_pozicije = cifre_parne_pozicije[::-1]  # √
    # provjera da li je prethodni kod ok
    # print(cifre_neparne_pozicije)
    # print(cifre_parne_pozicije)

    novi_broj = 0
    d = len(cifre_parne_pozicije)
    for i in range(d):
        novi_broj *= 10
        novi_broj += cifre_neparne_pozicije[i]
        novi_broj *= 10
        novi_broj += cifre_parne_pozicije[i]
    if d != len(cifre_neparne_pozicije):
        novi_broj *= 10
        novi_broj += cifre_neparne_pozicije[-1]
    return novi_broj

def is_prost(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def prosti_faktori(n):
    lista=[]
    for k in range(2, n+1):
        if(n % k == 0 and is_prost(k)):
            lista.append(k)
    return lista

if __name__ == '__main__':
    n = int(input("unesi broj: "))  # 15879249
    n1 = novi_broj(n)
    prosti_faktori_n = prosti_faktori(n)
    prosti_faktori_n1 = prosti_faktori(n1)
    proizvod1 = 1
    proizvod2 = 1

    for k in prosti_faktori_n:
        proizvod1 *= k

    for k in prosti_faktori_n1:
        proizvod2 *= k

    if (proizvod1 > proizvod2):
        print("Proizvod prostih faktora unesenog broja je veci od proizvoda prostih faktora novog broja")
    else:
        print("Proizvod prostih faktora unesenog broja nije veci od proizvoda prostih faktora novog \
              broja")
```

```python
# Zadatak 2

"""
Sa tastature se unosi tekst. Na osnovu unesenog teksta kreirati listu listi.
Elementi unutrašnje liste su riječi jedne rečenice. Npr. unesen je sljedeći tekst:
"Pozdrav! Kako si? Ja sam dobro, trenutno rjesavam zadatak iz programiranja."
Kreirana lista listi je [["Pozdrav"], ["Kako", "si"], ["Ja", "sam", "dobro", "trenutno",
"rjesavam", "zadatak", "iz", "programiranja"]].

Za početak cemo zanemariti zareze, navodnike, dvotačke... Koristićemo metodu `replace()`.
Zatim ćemo da razdvojiti tekst u rečenice.
Iz praktičnih razloga cemo koristiti [i1:i2], ali prvo trebamo odrediti indekse gdje su `.`, `!` i `?`.
Zatim razdvojimo rečenice u riječi, takođe koristimo metodu `split()`.
"""
if __name__ == '__main__':
    text = input("Unesite tekst: ")
    text = text.replace('"', "")
    text = text.replace("'", "")
    text = text.replace(",", "")
    text = text.replace(":", "")
    text = text.replace(";", "")
    text = text.replace("-", "")
    # sada smo "očistili" `text` od zareza, navodnika, ...
    # print(text)

    # indeksi `.`, `!` i `?`
    lista_indeksa = [i for i, c in enumerate(text) if c in ".!?"]  # елегантно тражење индекса! √
    # Da se prisjetimo kako funkcioniše uzimanje podstringa: `string[i1:i2]` je podstring stringa,
    # pri čemu je karakter na i1 poziciji uključen, a karakter na i2 nije.
    # Dakle, na početku liste indeksa trebamo dodati nula i na kraju dužinu `text`-a.

    if lista_indeksa[0] != 0:
        lista_indeksa.insert(0, -1)
    if lista_indeksa[-1] != len(text) - 1:
        lista_indeksa.append(len(text))

    print(lista_indeksa)

    # ako je lista_indeksa = [-1, 3, 6, 12, 14]
    # naš cilj je da uparimo sljedeće elemente: (0, 3), (4, 6), (7, 12), (13, 14)
    recenice = [text[lista_indeksa[i] + 1:lista_indeksa[i+1]] for i in range(0, len(lista_indeksa) - 1)]
    # print(recenice)

    lista_list = [recenica.split() for recenica in recenice]
    print("Trazena lista je: \n", lista_list)
```

```python
# Zadatak 4
# Sa tastature se unose brojevi sve dok se ne unese broj koji je veći od sume dva prethodno unesena broja.
# Kreirati listu listi prostih djelioca unesenih brojeva. Broj unesenih brojeva mora biti bar tri.

def prost(n):
    broj_je_prost = True
    for i in range(2, n):
        if n % i == 0:
            broj_je_prost = False
            break
    return broj_je_prost

def prosti_djelioci(n):
    lista = list()
    for i in range(2, n+1):
        if n % i == 0 and prost(i):
            lista.append(i)

    # prethodni kôd možemo zamijeniti sa
    # lista = [i for i in range(2, n+1) if n % i == 0 and prost(i)]  # одлично! √

    return lista

if __name__ == '__main__':
    lista_brojeva = list()  # иницијализација празне листе list() конструктором √
    b1 = int(input("Unesite prvi broj: "))
    lista_brojeva.append(b1)
    b2 = int(input("Unesite drugi broj: "))
    lista_brojeva.append(b2)
    b3 = int(input("Unesite treci broj: "))
    lista_brojeva.append(b3)

    while b1 + b2 >= b3:
        b1, b2 = b2, b3  # b1 = b2; b2 = b3;  # прекрасно! √
        b3 = int(input("Unesite sljedeci broj: "))
        lista_brojeva.append(b3)

    print("Unesena lista brojeva je:", lista_brojeva)

    lista_listi = [prosti_djelioci(k) for k in lista_brojeva]
    print("Lista listi prostih djelioca unesenih brojeva je:", lista_listi)
```

```python
# Zadatak 5
"""Sa tastature se unosi lista cijelih brojeva. Unos se prekida kada se unese nula.
Ispitati da li je proizvod cifara brojeva iz liste koji su palindrom jednak zbiru brojeva iz
liste koji nisu palindrom. Npr. [26262, 10, 200, 11, 78] -> true, [2, 34, 66, 10] -> false.
"""

lista = []
palindromi = []
nisupalindromi = []

def palindromf(broj):
    d = 0
    cifre = []
    t = int(broj)
    brojac = 0
    while t > 0:
        brojac = brojac + 1
        t = t // 10
    t = int(broj)
    for k in range(brojac):
        cifre.append(t % 10)
        t = t // 10
    z = brojac
    for k in range(brojac // 2):
        if cifre[k] == cifre[z-1-k]:
            continue
        elif cifre[k] != cifre[z-1-k]:
            d = d + 1
    return d

def palinprocif(broj):
    proizvod = 1
    cifre = []
    t = int(broj)
    brojac = 0
    while t > 0:
        brojac = brojac + 1
        t = t // 10
    t = int(broj)
    for k in range(brojac):
        cifre.append(t % 10)
        t = t // 10
    for k in cifre:
        proizvod = proizvod * k
    return proizvod

x = int(input())
while x != 0:
    lista.append(x)
    x = int(input())
for k in lista:
    d = palindromf(k)
    if (d == 0):
        palindromi.append(k)
    else:
        nisupalindromi.append(k)
zbir = 0
for k in nisupalindromi:
    zbir = zbir + k
proizvod = 1
for k in palindromi:
    proizvod = proizvod * palinprocif(k)
if (zbir == proizvod):
    print("true")
else:
    print("false")
```
